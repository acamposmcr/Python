```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('merge_from_ofoct.txt',names=["cita_nac","ced_padre", "ced_madre","cod_hosp","hora_nac","fecha_nac","sexo","relleno","nac_hijo","defunc","pais_padre","pais_madre","advert","ape1","ape2","nombre","nombre_padre","nombre_madre","lugar_nac","tipo_mov","fecha_marg","hora_marg","fecha_natura","indic_ext","prov_cant_madre","fec_aplic"],encoding="latin-1")
df_ene = pd.read_csv('merge_from_ofoct.txt',names=["dato"],encoding="latin-1")
```

```
df_ene.info
```

```
df["cita_nac"]=(df_ene["dato"].apply(lambda x: x[:12]))
df["ced_padre"]=(df_ene["dato"].apply(lambda x: x[12:21]))
df["ced_madre"]=(df_ene["dato"].apply(lambda x: x[21:30]))
df["cod_hosp"]=(df_ene["dato"].apply(lambda x: x[30:33]))
df["hora_nac"]=(df_ene["dato"].apply(lambda x: x[33:37]))
df["fecha_nac"]=(df_ene["dato"].apply(lambda x: x[37:45]))
df["sexo"]=(df_ene["dato"].apply(lambda x: x[45:46]))
df["relleno"]=(df_ene["dato"].apply(lambda x: x[46:48]))
df["nac_hijo"]=(df_ene["dato"].apply(lambda x: x[48:49]))
df["defunc"]=(df_ene["dato"].apply(lambda x: x[49:50]))
df["pais_padre"]=(df_ene["dato"].apply(lambda x: x[50:53]))
df["pais_madre"]=(df_ene["dato"].apply(lambda x: x[53:56]))
df["advert"]=(df_ene["dato"].apply(lambda x: x[56:57]))
df["ape1"]=(df_ene["dato"].apply(lambda x: x[57:70]))
df["ape2"]=(df_ene["dato"].apply(lambda x: x[70:83]))
df["nombre"]=(df_ene["dato"].apply(lambda x: x[83:113]))
df["nombre_padre"]=(df_ene["dato"].apply(lambda x: x[113:142]))
df["nombre_madre"]=(df_ene["dato"].apply(lambda x: x[142:171]))
df["lugar_nac"]=(df_ene["dato"].apply(lambda x: x[171:200]))
df["tipo_mov"]=(df_ene["dato"].apply(lambda x: x[200:201]))
df["fecha_marg"]=(df_ene["dato"].apply(lambda x: x[201:209]))
df["hora_marg"]=(df_ene["dato"].apply(lambda x: x[209:215]))
df["fecha_natura"]=(df_ene["dato"].apply(lambda x: x[215:223]))
df["indic_ext"]=(df_ene["dato"].apply(lambda x: x[223:224]))
df["prov_cant_madre"]=(df_ene["dato"].apply(lambda x: x[224:227]))
df["fec_aplic"]=(df_ene["dato"].apply(lambda x: x[227:235]))


#, "","","","","","","","","","","","","","","","","","","","","","",""
```
